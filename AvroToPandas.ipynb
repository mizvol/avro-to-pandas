### Download '.avro' files and convert them to Pandas DataFrame

```python
# pip install pandavro
import pandavro as pdx # 'avro' to 'pandas' library.
import urllib.request # download files
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
base_url = "https://revo-prod-data.s3.amazonaws.com/data/"
file_name = "25c2a84d-6589-41a3-b731-d5bdce2aee74.avro"
```

Download file and save it in the rood directory

```python
urllib.request.urlretrieve(base_url + file_name, file_name)
```

Read '.avro' file and convert it to Pandas DataFrame

```python
df = pdx.from_avro("./" + file_name)
```
